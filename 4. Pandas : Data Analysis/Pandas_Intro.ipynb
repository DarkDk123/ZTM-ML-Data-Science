#### *loc* & *iloc* functions to select Data : 
* ***.loc[]***  is used to select data using **Indexes**.
* ***.iloc[]***  is used to select data using **Positions** [0 to length-1] (Like python lists).

```python
# Creating a Series.

snake_series = pd.Series(
    data=["Python", "King Cobra", "Bush Viper", "Hog Nosed", "Russel's Viper"],
    index=[1, 1, 2, 4, 3]  # Custom Indexes acc. to favorites.
)
snake_series
```

```python
## Using .loc to select using indexes.
snake_series.loc[1] # Both python & King Cobra has index "1".
```

```python
## Using .iloc to select using Positions i.e. "0, 1, 2,..."
print(snake_series.iloc[1]) ## King Cobra
print(snake_series.iloc[4]) ## Position 4 = Russel's Viper
print(snake_series.loc[4])  ## Index 4 = Hog Nosed
```

### Slicing in *loc* & *iloc* methods :
Same as Python lists we can use **Slicing** to get a range of index.

```python
## Slicing
snake_series.iloc[:3] # Slice positions from start to 3(excluding)
```

```python
## Slicing using indexes.
snake_series.loc[:3] ## all indexes from start to first occurrence of 3(including).
```

#### **Importing Data with Missing Values :**

* We'll use *`inplace`* parameter to update Data inplace without the need of reassignment.

```python
df_missing = pd.read_csv("./car-sales-missing-data.csv")
df_missing ## DataFrame with missing values ("NAN")
```

```python
# Fill Odometer `NAN` values with 'mean' of Odometer.
# Using the `inplace` argument.
df_missing["Odometer"].fillna(
    round(df_missing["Odometer"].mean(), 2), # Rounding for better formatting]
    inplace = True ## Modifying inplace
    ) 
```

```python
df_missing ## Filled Missing Values.
```

#### **Dropping rows/cols containing Missing Values :**
* using *DataFrame.dropna()* method.

```python
 # Dropping all rows containing NaN values.
df_missing.dropna(inplace=True) # `inplace` will make all the dropped rows 'inaccessible'.
```

```python
print("Dropped Missing Column :\n\n", df_missing)
# Exporting Cleaned DataFrame in Storage.
df_missing.to_csv("./cleaned_car-sales-missing.csv", index=False)
```

#### **Creating new Columns :**
1. From ***Pandas.Series*** object (can be of varying length).
2. From ***Python List,Tuple etc.*** Objects (Must be of same length as DataFrame)

```python
## Adding column using a Pandas.Series object
df_imported["Seats"] = pd.Series([3, 5, 4, 2, 5])
df_imported ## we have missing values | Size of added series wasn't same.
```

```python
# we can fill NaN values in "Seats" Column
df_imported["Seats"].fillna(4, inplace=True)
df_imported # Filled NaN values.
```

```python
# Creating new column using "Python Lists"
#df_imported["Fuel per 100 KM"] = [3.8, 4.2, 7.1, 8.9, 5.5, 6.4] # ValueError: Length doesn't match.
df_imported["Fuel per 100 KM"] = [3.8, 4.2, 7.1, 8.9, 5.5, 6.4, 7.7, 9.4, 3.4, 4.7]
df_imported 
```

```python
# Creating new col with other columns
# Creating column containing "Total Fuel Used".
df_imported["Total Fuel Used"] = df_imported["Odometer (KM)"]*df_imported["Fuel per 100 KM"]/100

# updated DataFrame
df_imported
```

```python
## Renaming a Column
df_imported.rename(
    columns={
        'Fuel per 100 KM': 'Fuel per 100 KM (L)',
        'Total Fuel Used':'Total Fuel Used (L)'
    },
    inplace=True
    )
df_imported
```

```python
# Create a Column Directly from single value ("Scaler multiply in Linear Algebra").
df_imported["Is Safe"] = True
df_imported
```

### **Some other Functions.**

```python
## `Sample` method to get fraction of whole Data.
df_imported.sample(frac=0.5) # 50% Of entire Data (frac ranges from 0 to 1)
```

```python
# Getting entire DataFrame Shuffled (frac=1)
shuffle_df_imported = df_imported.sample(frac=1)
shuffle_df_imported
```

```python
shuffle_df_imported.reset_index(drop=True, inplace=True) # reset index of shuffled DataFrame
shuffle_df_imported
```

```python
# DataFrame.apply() method to apply any specific function to the Data of Our DataFrame.
shuffle_df_imported["Odometer (KM)"] = shuffle_df_imported["Odometer (KM)"].apply(lambda x: x/1.6)
shuffle_df_imported
```
