```python
# Allow Matplotlib figures to be displayed on Jupyter Notebook.
# %matplotlib inline # jupyter's magic command

# import required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## **`matplotlib.pyplot` | 1st Way:**

* pyplot API plots in a state-less plotting.
* It is generally less-flexible .

```python
# Simplest plot | Line Plot using plt.plot() | plot x Vs y as lines/markers
plt.plot() # No-data | empty figure
plt.show() # show our plot
```

```python
# Let's plot a 1D Data
plt.plot([2, 4, 6, 8, 10]) # single-axis value
plt.show()
```

```python
# Let's create X Vs y plot.
x = [2, 4, 6, 8, 10]
y = [1, 3, 5, 7, 9] # Called State-less plotting 

plt.plot(x, y) # Two-axes
plt.show()
```
