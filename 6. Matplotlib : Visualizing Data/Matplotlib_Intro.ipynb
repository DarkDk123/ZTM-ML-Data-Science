## **`matplotlib.pyplot` | Stateful approach | 1st Way:**

- pyplot API plots in a stateful plotting.
- It is generally less-flexible .
- It works ever with current figure & current axes only.
- Directly pyplot functions are used for this **like** `plt.plot`, `plt.ylabel` etc.


```python
# Simplest plot | Line Plot using plt.plot() | plot x Vs y as lines/markers
plt.plot() # No-data | empty figure
plt.show() # show our plot
```

```python
# Let's plot a 1D Data
plt.plot([2, 4, 6, 8, 10],) # single-axis value 
plt.show()
```

```python
# Let's create X Vs y plot.
x = [2, 4, 6, 8, 10]
y = [1, 3, 5, 7, 9] # Called State-less plotting 

plt.plot(x, y) # Two-axes
plt.show()
```

```python
# Another Example of Stateful approach of plotting.
plt.scatter(x, y)
plt.xlabel("X_Axis") # Set x-label of current Axes
plt.ylabel("Y_Axis") # Set y-label.
# Setting x-axis & y-axis limits
plt.ylim(0, 12)
plt.xlim(0, 12)
plt.show() # show current Figure. # all figures are lost here!
```

```python
# (It's Complex) I have worked with it so let's do!!

# Stateful approach | Single Figure with Multiple Axes.
# Create the first subplot
plt.subplot(2, 2, 1)
plt.scatter([1, 2, 3], [4, 5, 6])
plt.title("Subplot 1")

# Create the second subplot
plt.subplot(2, 2, 2)
plt.plot([1, 2, 3], [2, 4, 6])
plt.title("Subplot 2")

# Create the third subplot
plt.subplot(2, 2, 3)
plt.bar([1, 2, 3], [5, 3, 7])
plt.title("Subplot 3")

# Create the fourth subplot
plt.subplot(2, 2, 4)
plt.plot([1, 2, 3], [8, 1, 4])
plt.title("Subplot 4")

# Title to Figure 
plt.suptitle("'Stateful Plot'\nFigure with Multiple Axes", fontdict={"size":15, "weight":"bold", "color": "Orange"})

# Adjust spacing between subplots
plt.tight_layout()

# Show the figure
plt.show()
```

## **`matplotlib` | Object-Oriented Approach | 2nd Way:**

- In the object-oriented approach, you create explicit figure and axes objects.
- This method provides more control over layout and customization.
- Each axes object corresponds to a subplot and can be customized individually.
- Customization of elements like labels and titles is done using dedicated methods.
- Preferred for complex layouts and multi-panel visualizations.

**Let's see multiple ways to do `Object Oriented Plotting` with Matplotlib.**


```python
# 1st Method
fig = plt.figure() # Create new Figure obj
ax = fig.add_subplot() # Add axes obj in it.
plt.show() # show
```

```python
# 2nd Method
fig = plt.figure() # Create new figure obj
ax = fig.add_axes(rect=(.7, .7, .7, .7)) # Create new axes obj  ## with given size
ax.plot(x, y)
plt.show()
```

```python
# 3rd Method (Recommended by instructor)
fig, ax = plt.subplots() # Axes with Figure objects
ax.plot(x, y); # Plotting (";" for showing appropriately)
```

## **Matplotlib Workflow :**

- This general steps are used to Visualize data in a Project!!


```python
# 1. Importing matplotlib & getting ready to visualize in Jupyter.
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 2. Preparing Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 3. Creating figure & axes.
fig, axes = plt.subplots(figsize=(10, 7))

# 4. Plotting data.
axes.scatter(x, y,
            color='w', edgecolor='b'
        )
# 5. Customizing the Plot
axes.set(
    xlabel='x',
    ylabel='sin(x)',
    title='Title - sin(x) vs. x',
)
axes.title.set_fontsize(20) # Size of Title.

# 6. save the Figure
fig.savefig("Images/scatter.png")
```

### **Two Different Options for Subplot :**
    Multiple Axes objects within the same figure.

1. Directly Unpacking into different axes objects.
2. Storing as an Array in a Variable, then using via grid i.e. `ax[0, 0], ax[0, 1]` etc.

```python
# Option 1

# Single Figure wit 2 Axes (graphs) | 1 row & 2 column.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plotting random dummy data.

# First Plot | Line plot
ax1.plot(x.cumsum(), "g-", linewidth=4.5)
ax1.set(
    title="Line Plot",
    xlabel="x-axis",
    ylabel="y-axis",
)

# Second Plot | Histogram plot
ax2.hist(np.random.randn(1000))
ax2.set(
    title="Histogram Plot",
    xlabel="x-axis",
    ylabel="y-axis",
)

# Display Figure
fig.suptitle("Subplots Figure  | Multiple Axes", size=16, fontdict={'weight':'bolder', 'color':'greenyellow'})
fig.tight_layout()
plt.show()
```

```python
# Option 2

# Single Figure wit 2 Axes (graphs) | 1 row & 2 column.
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Plotting random dummy data.

# First Plot | Line plot
ax[0].plot(x.cumsum(), "g-", linewidth=4.5)
ax[0].set(
    title="Line Plot",
    xlabel="x-axis",
    ylabel="y-axis",
)

# Second Plot | Histogram plot
ax[1].hist(np.random.randn(1000))
ax[1].set(
    title="Histogram Plot",
    xlabel="x-axis",
    ylabel="y-axis",
)

# Display Figure
fig.suptitle("Subplots Figure  | Multiple Axes", size=16, fontdict={'weight':'bolder', 'color':'greenyellow'})
fig.tight_layout()
plt.show()
```

Pandas uses Matplotlib at the backend, to create these visualizations, so we can use all the matplotlib objects, methods etc. to customize our graphs.

```python
## Creating a Random Series to visualize it's cumulative sum. (numpy.random | seed=27)
np.random.seed(27)

random_series = pd.Series(
    np.random.randn(100),
    index= pd.date_range("6 september 2023", "6 september 2024", periods=100) # DateTime index.
    )

cumsum = random_series.cumsum() # cumulative sum
ax = cumsum.plot(color='yellow', marker='.'); # Plot a Line Graph.

date_format = plt.matplotlib.dates.DateFormatter('%d %b %Y') # matplotlib date formatter.
ax.xaxis.set_major_formatter(date_format)

plt.show()
```

#### **Plotting From Pandas DataFrame :**

Using Car_sales CSV File.

```python
## Let's View Car_sales DataFrame Again !!
car_sales
```

```python
## Again, Like we have seen before!
## Check the Information of the DataFrame.
## Check Data Types of each columns. (They are appropriate or not?)
car_sales.info() # Price col must be in Integer DataType.
```

```python
# Convert data type of `Price` column using RegEx.
car_sales.Price = car_sales.Price.str.replace(
                "[\D]",
                "",
                regex=True
        ).str.removesuffix("00").astype(int)

car_sales.Price ## Converted correctly to int.
```

```python
## Now, Let's add a new column named `sale_date`.

car_sales["Sale_Date"] = pd.date_range(
    "29 November 2022",
    freq="7D",  # Frequency of 7 days
    periods=len(car_sales),  # Date Series of the same size as the DataFrame
)

# Format the Sale_Date column
car_sales["Sale_Date"] = car_sales["Sale_Date"].dt.strftime('%d %b %Y')

car_sales ## We got it!! The Sale_Date Column.
```

```python
# Now, add a `Total_Sum` column having a cumulative sum of the `Price` column.
car_sales["Total_Sum"] = car_sales.Price.cumsum()
car_sales
```

```python
# Let's plot the Total_Sales over time.
ax = car_sales.plot(x="Sale_Date", y="Total_Sum", linestyle='--', marker='o')

# set axis labels.
ax.set(
    xlabel="TIME",
    ylabel="Total Sales"
)

# Customize Axes title.
ax.set_title(
    label = "Increasing Sales Over Time\n",
    fontdict={
        'color':'yellowgreen',
        'size': 16,
        'weight':'bold'
    }
)

ax.tick_params(rotation=30, labelcolor='gold') # Rotate ticks for better representation.
ax.get_legend().remove() # Removing the legend!

plt.show()
```

```python
## plotting scatter plot
ax = car_sales.plot(
    x="Odometer (KM)",
    y="Price",
    kind="scatter",
    edgecolor="blue",
    color='red'
)

# set axis labels.
ax.set(
    xlabel="Odometer KM",
    ylabel="Price"
)

# Customize Axes title.
ax.set_title(
    label = "Price based on Odometer Value\n",
    fontdict={
        'color':'brown',
        'size': 16,
        'weight':'bold'
    }
)

ax.tick_params('x', rotation=30, ) # Rotate ticks for better representation.

plt.show()
```
